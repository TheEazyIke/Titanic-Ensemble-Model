Report:

- Dataset Summary: We identified that the dataset had 891 and 12 columns which PassengerId, Survived, Pclass, Name, Sex, Age, SibSp, Parch, Ticket, Fare, Cabin, and Embarked. Of these, Age and Fare were continuous variables and Name, Sex, Embarked, Cabin, and Ticket were nominal categorical variables. These nominal categorical variables will either need to be encoded during feature engineering, or dropped if they have too many unique values, as is the case for Name, Cabin, and Ticket. The Sex column was discovered to have only 2 classes (male or female), so we can turn it into a binary dummy variable during feature engineering. The other columns, Pclass, SibSp, and Parch were ordinal categorical variables. Lastly, we discovered that the distribution of the Survived column was imbalanced, with only 38% of records indicating survival.
  
- Missing Value Treatment: We identified that Age, Cabin, Embarked columns have null values needing to be imputed. Both Age and Embarked had few enough missing values that we were able to perform imputation. For the Age column we used SimpleImputer with the median strategy, since it is a continous variable. For the Embarked column we used the SimpleImputer with the most_frequent strategy (mode), as it is a nominal categorical column. However, the Cabin column actually has so many that it will need to be dropped after using it for feature engineering, as the data would not be reliable given that it is over 75% null.

- EDA: EDA exposed a few important data considerations. First off, we discovered that men were much more likely to die, as only 32% of survivors were men, while 85% of deaths were men. Next we discovered that although survivors tended to be about the same age, nonsurviving females tended to be younger compared to nonsurviving males. We also exposed that the Age and Fare columns had many outliers, with Age having 66 and Fare having 222. Addtionally, we discovered that the Passenger's Class greatly impacted survival rates. Roughly 65% of 1st class passengers survived, while roughly 45% of 2nd class passengers survived, and only about 25% of 3rd class passengers survived. We also discovered a positive correlation between SibSp and Parch and a negative correlation between Fare and Pclass, providing us with important important information for feature creation later on. Lastly, we discovered that Fare has a right skewed distribution, meaning we will need to perform a transformation on during feature engineering, in order to make it more Gaussian/Normal.

```python
#Family feature using SibSp and Parch since correlation matrix showed significant correlation (about 0.5)
#interaction features: family (sibsp + parch), floor_level based on first part of cabin column

from sklearn.preprocessing import PolynomialFeatures

pf = PolynomialFeatures(degree=2, interaction_only=False, include_bias=False)

#creating floor level feature
df['Floor_Level'] = df.Cabin.str[0:1]
#filling with n to represent 
df.Floor_Level.fillna(value='N', inplace=True)
df_oh = pd.DataFrame(oh_encoder.fit_transform(df[['Floor_Level']]))
df = df.join(df_oh)
df.drop(['Floor_Level'], axis=1, inplace=True)

#Creating HasCabin feature
df['HasCabin'] = df.Cabin.apply(lambda x: 0 if pd.isnull(x) else 1)

#dropping cabin column because it's no longer needed and has too many null values
df.drop(['Cabin'], axis=1,inplace=True)

#creating family feature based on the correlation between sibsp and parch (both contain loved ones you may be with)
df['Family'] = df.SibSp + df.Parch
```

```python
#applying PolynomialFeature function to capture interactions between existing and newly created variables 
    #Example: Pclass and Fare, which had a strong negative correlation in the correlation matrix at the beginning
#also want to identify non linear relationships between target variable and features like Age or Fare
    #example: Babies or elderly may be much more likely to have survived, as their survival may be prioritized by others
df_y = df['Survived']
df_x = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare',
       'Embarked_Q', 'Embarked_S', 'Floor_Level_B', 'Floor_Level_C',
       'Floor_Level_D', 'Floor_Level_E', 'Floor_Level_F', 'Floor_Level_G',
       'Floor_Level_N', 'Floor_Level_T', 'HasCabin', 'Family']]
poly_features = pf.fit_transform(df_x)
feature_names = pf.get_feature_names_out(input_features=df_x.columns)

poly_df = pd.DataFrame(poly_features, columns = feature_names)

df_with_poly = pd.concat([df_x, poly_df], axis=1)
```

```python
from sklearn.decomposition import PCA

#applying because:
#SVM would overfit from too much noise due to large feature space
#Naive Bayes assumes all features are independent, meaning error increases with every additional feature
#Although not strictly necessary for Random Forest, it may help reduce computation time
pca = PCA(n_components=0.95) #enough features to retain 95% of variance

pca_x = pca.fit_transform(df_x)

n_components = pca.n_components_  # Number of principal components
column_names = [f'PC{i+1}' for i in range(n_components)]

explained_variance = pca.explained_variance_ratio_
print("Explained variance by components: ", explained_variance)

df_pca_x = pd.DataFrame(pca_x, columns=column_names)
```
